# ***Binomial and Uniform Distribution.***

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
n,p = 10,0.5
s1 = np.random.binomial(n,p,1000)   #Sample Space

x = []
```

```python
for i in range(1000):
  x.append(random.uniform(0,10))
```

```python
count,bins,ignore = plt.hist(s1,10,align='mid',rwidth=0.9,ec='black',label='Binomial Plot.')
count,bins,ignore = plt.hist(x,10,align='mid',rwidth=0.8,ec='black',histtype='step',hatch='/',label='Uniform Distribution Plot.')

plt.legend(loc='best')
plt.xlabel('$Bins$ -------->')
plt.ylabel('$Frequency$------->')
plt.grid()
plt.show()
```
