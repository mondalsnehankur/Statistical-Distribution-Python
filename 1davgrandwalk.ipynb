# ***1-D Average Random Walk***

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def pwr(x,a,b):
  return(a*x**b)

steps = 200
girls = 2000

x = np.zeros([steps])
t = range(steps)
x_sum = np.zeros([steps])
x2_sum = np.zeros([steps])
```

```python
for j in range(1,girls):
  for i in range(1,steps):
    if(random.choice([0,1])==0):
      x[i] = x[i-1] - 1
    else:
      x[i] = x[i-1] + 1
    x_sum[i] += x[i]
    x2_sum[i] +=x[i]**2

#print(x)
x_avg = [float(i)/girls for i in x_sum]
RMS = [np.sqrt(float(i)/girls) for i in x2_sum]
```

```python
plt.plot(t,x_avg,label='Average Walk Plot.')
plt.plot(t,RMS,label='RMS Plot.')

par,var = curve_fit(pwr,t,RMS)
plt.plot(t,pwr(t,par[0],par[1]),label='Curve Fit.')

plt.text(125,-0.7,'--Average RMS Plot.')
plt.text(79,8,'--RMS Plot.')
plt.text(30,8,'Curve Fit.--')

plt.xlabel('$Time$ $(step number)$ ------>')
plt.ylabel('$Average$ & $RMS$ $position(steps)$ ------->')
plt.grid()
plt.legend(loc='best')
plt.title('1-D Average Random Walk.')
plt.axhline(0,color='black',linestyle='--')
plt.axvline(0,color='black',linestyle='--')

plt.show()

print("Power Fit : y(t) = A x t^B")
print("A = %f +/- %f ."%(par[0],np.sqrt(var[0,0])))
print("B = %f +/- %f ."%(par[1],np.sqrt(var[1,1])))
```
