# ***Plotting Uniform Distribution and henceforth converting it to Exponential Distribution.***

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
N = 10000

u = np.random.uniform(0,1,size=N)   # Uniform distribution: Random Number.
```

```python
plt.hist(u,bins=10,ec='black')
plt.title("Uniform Distribution.")
plt.ylabel("$N$------>")
plt.xlabel("$bin$------>")
plt.grid()
```

**Conversion from Uniform Distribution to Exponential Distribution.**

```python
N = 10000

u = np.random.uniform(0,1,size=N)   # Uniform distribution: Random Number.
x = -np.log(1-u)    # Mapping Rule to Exponential From

freq,binv,extra = plt.hist(x,bins=20)
bin_mid = (binv[:-1]+binv[1:])/2.0

print(binv,'\n\n',binv[:-1],'\n\n',binv[1:],'\n\n\n')

binv_width = binv[1]-binv[0]

print("Width : ",binv_width,'\n\n')

from_theory = np.exp(-bin_mid)*binv_width*N

# PLOTTING
plt.plot(bin_mid,from_theory,'-o',label="Mid Point")
plt.legend(loc='best')
plt.title("$Exponential$")
plt.xlabel("$bin$ ---------->")
plt.ylabel("$Frequency$ ---------->")
plt.grid()
plt.show()
```
