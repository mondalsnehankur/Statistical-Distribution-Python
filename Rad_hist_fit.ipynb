# ***Radioactive Histogram Fit***

```python
import numpy as np
from scipy.stats import norm
from numpy import random
import matplotlib.pyplot as plt
```

```python
Trials = 1000
N_d = []
for n in range(Trials):
  N_decay = 0
  N = 500
  Nleft = N
  lam = 0.01
  t = 100
  dt = 10
  for t in range(0,t,dt):
    for i in range(Nleft):
      x = random.random()
      if(x<=lam*dt):
        Nleft-=1
        N_decay+=1
  N_d.append(N_decay)
```

```python
(mu,sigma) = norm.fit(N_d)
n,bins,patches = plt.hist(N_d,bins=20,density=1,rwidth=0.9,ec='black',label='Histogram Plot.')

y = norm.pdf(bins,mu,sigma)
plt.plot(bins,y,lw=2,label='Normalised Function.')

#plt.plot(x,np.exp(-(x-mu)**2/(2*std))/np.sqrt(2*np.pi*std),label='Function Plot.')

plt.xlabel('No:of Counts ------>')
plt.ylabel('$Frequency$ ----- >')
plt.legend(loc='best')
plt.grid()
plt.show()
```
