# ***Biased Coin.***

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
N = 1000  # Trials
n=100     # Samples
p=0.1     # Setting the biasness to 0.1

result = []
prob = np.zeros(n+1)
```

```python
for j in range(N):
  success = 0
  for i in range(n):
    toss = random.random()
    #print(toss)
    if(toss <= p):
      success += 1
  #print(success)
  result.append(success)
  prob[success] += 1

x = np.arange(0,n+1)
```

```python
#Plotting

plt.hist(result,n,ec='black',color='orange')
plt.title("n=10 ,p=0.1, No:of trials = 1000")
plt.grid()
plt.ylabel("Count ------>")
plt.xlabel("$n$ ------>")
plt.xlim(-5,105)

plt.show()
```

```python
plt.plot(x,prob,"o",ms=10,label="Total Success.")
plt.legend(loc="best")
plt.title("Bin ( 10 , 0.1 )")
plt.grid()
plt.ylabel("Probability ($p(x)$) ------>")
plt.xlabel("$x$ ------>")

plt.show()
```
