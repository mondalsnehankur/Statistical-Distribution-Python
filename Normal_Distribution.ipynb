# ***Normal ( Gaussian ) Disttribution.***

*CLT (Central LImit Theoram) states that regardless of the variable's distribution in the population, the sampling distribution of the mean will tend to approximate the normal distribution.*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Initialisation

n = 100000
population = np.random.uniform(0,1,size=n)

mean_sample = []
```

```python

for x in range(1,n):
  sample = np.random.choice(population,size=int(0.01*len(population)))
  mean_sample.append(np.mean(sample))

mean_meansamp = np.mean(mean_sample)

var_meansamp = np.var(np.mean(mean_sample))

mean_population = np.mean(population)

var_population = np.var(population)
```

```python
# OUTPUT
```

```python
print("Mean of sample means : ",mean_meansamp,'\n')
print("Mean of population : ", mean_population,'\n')
print("Variance of sample means : ",100*var_meansamp,'\n')
print("Variation of population : ",var_population,'\n')

plt.hist(mean_sample,30,color='red',ec='black')

plt.grid()
plt.xlabel("$bin$ ---------->")
plt.ylabel("$Frequency$ ---------->")
plt.title(" $Normal Distribution.$ ")

plt.show()
```
