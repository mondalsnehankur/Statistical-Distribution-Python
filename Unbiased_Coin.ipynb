# ***Unbiased Coin.***

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
N = 1000
n=10
p=0.5

result = []
prob = np.zeros(n+1)
```

```python
for j in range(N):
  success = 0
  for i in range(n):
    toss = random.random()
    #print(toss)
    if(toss <= p):
      success += 1
  #print(success)
  result.append(success)
  prob[success] += 1

x = np.arange(0,n+1)
```

```python
plt.hist(result,n,ec='black',color='orange')
plt.title("n=10 ,p=0.5, No:of trials = 1000")
plt.grid()
plt.ylabel("Count ------>")
plt.xlabel("$n$ ------>")

plt.show()
```

```python

plt.plot(x,prob,"bo",ms=10,label="Total Success.")
plt.legend(loc="best")
plt.title("Bin ( 10 , 0.5 )")
plt.grid()
plt.ylabel("Probability ($p(x)$) ------>")
plt.xlabel("$x$ ------>")

plt.show()
```
