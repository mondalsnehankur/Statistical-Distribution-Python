# ***Binomial Distribution of Unbiased Tossed of Coin.***

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
n,p = 10,0.5
s1 = np.random.binomial(n,p,10000)

count,bins,ignore = plt.hist(s1,10,align='mid',rwidth=0.9,color='grey',ec='black',label='Binomial Distribution Plot.')
#print s1   # Sample Space

heads = []

for i in range(10000):
  tosses = [np.random.random() for i in range(n)]
  heads.append(len([j for j in tosses if j>=p]))

count,bins,ignore = plt.hist(heads,10,align='mid',rwidth=0.9,color='grey',ec='black',histtype='step',hatch='/',label='Uniform Distribution Plot.')

plt.legend(loc='best')
plt.ylim(-50,3000)
plt.xlabel('$No:of heads$ ------>')
plt.ylabel('$Frequency$ --------->')
plt.axhline(0,color='black',linestyle='--')
plt.axvline(0,color='black',linestyle='--')
plt.grid()
plt.show()
```
